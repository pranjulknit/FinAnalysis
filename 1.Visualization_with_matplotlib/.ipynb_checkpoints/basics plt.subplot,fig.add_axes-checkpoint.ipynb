```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y)
#plt.show()
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
```

```python
plt.subplot(1,2,1)
plt.plot(y,x,'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
axes.set_xlabel('xlabel')
axes.set_ylabel('ylabel')
axes.set_title('settitle')
axes.plot(x,y)
```
